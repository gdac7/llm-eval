```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

# Dados para o gráfico
categorias = ['Construção', 'Regulação', 'Transferência']
acuracias = [69.6, 75.4, 67.5]  # Médias das categorias

# Cálculo do intervalo de confiança
confidence_level = 0.95
intervalos = []

for i, mean in enumerate(acuracias):
    # Dados hipotéticos para cálculo do intervalo
    # Substitua `vec` pelo vetor de dados reais de cada categoria
    vec = np.random.normal(mean, 5, 30)  # Simula dados ao redor da média
    intervalo = st.t.interval(
        confidence_level, len(vec) - 1, loc=np.mean(vec), scale=st.sem(vec)
    )
    intervalos.append(intervalo)

# Calculando barras de erro (yerr)
yerr = [
    (mean - intervalo[0], intervalo[1] - mean) for mean, intervalo in zip(acuracias, intervalos)
]
erro_superior = [e[1] for e in yerr]
erro_inferior = [e[0] for e in yerr]
# Criar gráfico de barras com barras de erro
fig = plt.figure(figsize=(8, 6))
ax = fig.gca()
ax.bar(categorias, acuracias, color="blue", yerr=erro_superior, capsize=5)

# Configurações do gráfico
#plt.title('Acurácia por Categoria', fontsize=15)
plt.xlabel('Categorias', fontsize=15)
plt.ylabel('Acurácia (%)', fontsize=15)
plt.ylim(0, 100)  # Limite de 0 a 100 para a acurácia
plt.grid(linestyle='dotted', axis='y')

plt.tick_params(labelsize=14)
plt.subplots_adjust(left=0.2, bottom=0.18, right=0.9, top=0.888, wspace=0.2, hspace=0.2)

# Ajustar a espessura das bordas
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2)

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
categorias = ['Construção', 'Regulação', 'Transferência']
modelo1 = [81.63, 77.66, 76.46]
modelo2 = [68.93, 83.75, 75.46]
modelo3 = [82.75, 77.45, 80.79]

cores = ['#FFB6C1', '#87CEEB', '#98FB98']

x = np.arange(len(categorias))  # Localizações dos grupos
largura = 0.2  # Largura das barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - largura, modelo1, largura, label='Llama-fp16', color=cores[0])
rects2 = ax.bar(x, modelo2, largura, label='Llama-8bits', color=cores[1])
rects3 = ax.bar(x + largura, modelo3, largura, label='Llama-4bits', color=cores[2])

# Adiciona texto, rótulos, etc.
ax.set_xlabel('Categorias')
ax.set_ylabel('Acurácia (%)')
ax.set_xticks(x)
ax.set_xticklabels(categorias)
ax.set_ylim(0, 100)

# Posiciona a legenda horizontalmente acima do gráfico
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)

fig.tight_layout()

plt.show()
fig.savefig("acc_llama.pdf", bbox_inches='tight')
```
